```python
import numpy as np
import matplotlib.pyplot as plt

def plot_comparison(original_data, reconstructed_data, data_name):
    plt.figure(figsize=(10, 6))

    # Original plot
    plt.subplot(2, 1, 1)
    if original_data.shape[1] == 2:
        plt.plot(original_data[:, 0], label='Left Channel', color='red')
        plt.plot(original_data[:, 1], label='Right Channel', color='blue')
    else:
        plt.plot(original_data, label='Original', color='green')
    plt.xlabel('Samples')
    plt.ylabel('Amplitude')
    plt.title(f'Original {data_name}')
    plt.legend()
    plt.grid()

    # Reconstructed plot
    plt.subplot(2, 1, 2)
    if reconstructed_data.shape[1] == 2:
        plt.plot(reconstructed_data[:, 0], label='Left Channel', color='red')
        plt.plot(reconstructed_data[:, 1], label='Right Channel', color='blue')
    else:
        plt.plot(reconstructed_data, label='Decrypted', color='green')
    plt.xlabel('Samples')
    plt.ylabel('Amplitude')
    plt.title(f'Decrypted {data_name}')
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.show()


data_or = np.load("sound/data1.npy")
data_or_shape = data_or.shape
print(data_or_shape) 

data_1 = np.load("decrypted_sound/decrypted_sound_7.npy")
print("Before",data_1.shape)
data_1 = data_1[:data_or_shape[0],:]
print(data_1.shape)

# exit()
diff_indices = np.where(data_or != data_1)


# Print differences
if len(diff_indices[0]) > 0:
    print("Differences found at indices:")
    for i in range(len(diff_indices[0])):
        idx = tuple(d[i] for d in diff_indices)  # Get the index in multi-dimensional array
        print(f"Index {idx}: Original={data_or[idx]}, Decoded={data_1[idx]}")
else:
    print("No differences found!")

plot_comparison(data_or, data_1, "sound")
```
